```python
import torchaudio, os, IPython, torch
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import numpy as np
import librosa
import soundfile as sf
import os
```

```python
data_path = "../datasets/audioMinst/"
for file in os.listdir(data_path):
    if "wav" in file:
        y, sr = librosa.load("{}{}".format(data_path, file))
        y_slow = librosa.effects.time_stretch(y, rate=0.7)
        y_fast = librosa.effects.time_stretch(y, rate=1.3)
        sf.write('../datasets/audioMnistTimeStretch/{}_slow.wav'.format(file.split(".")[0]), y_slow, sr, 'PCM_24')
        sf.write('../datasets/audioMnistTimeStretch/{}_fast.wav'.format(file.split(".")[0]), y_fast, sr, 'PCM_24')
```
